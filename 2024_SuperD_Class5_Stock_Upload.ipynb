# Analyze return today and return tomorrow

```python
from scipy import interpolate
# Create column of Return with lag -1
df["ShiftReturn"] = df["Return"].shift(1)
df.dropna(inplace=True)
```

```python
# sort dataframe using Return in ascending order 
df.sort_values(by="Return",ascending=True,inplace=True)

# to make life easier
x = df["Return"]
y = df["ShiftReturn"]
```

```python
# Plot 
# set figure size
plt.figure(figsize=(8, 6))
# set title
plt.title("Return today vs Return tomorrow")
# plot scatter
plt.plot(x, y, '.', c="grey", label="original")

# set x label and y label
plt.xlabel("today",size=14)
plt.ylabel("tomorrow",size=14)
# add grid
plt.grid()
# show
plt.show() 
```

# Linear Regression

```python
import statsmodels.api as sm
# stats model
reg_linear = sm.OLS(y, x)
result = reg_linear.fit()
# prediction
y_linear = result.predict(x)
```

```python
# Plot: set figure size
plt.figure(figsize=(8, 6))
# set title
plt.title("Return today vs Return tomorrow")
# plot scatter
plt.plot(x, y, '.', c="grey", label="original")
# plot spline fit
plt.plot(x, y_linear, '-', c="g", label="linear fit")
# plot legend
plt.legend(loc='best', fancybox=True, shadow=True)
# set x label and y label
plt.xlabel("today",size=14)
plt.ylabel("tomorrow",size=14)
# add grid
plt.grid()
# show
plt.show() 
```

# Spline

```python
# number of nots
# Avoid error related to the Schoenberg-Whitney conditions
knot_numbers = 15

# we place nots using quantiles
x_quantile = np.linspace(0, 1, knot_numbers+2)[1:-1]
q_knots = np.quantile(x, x_quantile) 

x_min = np.quantile(x,0.001)
x_max = np.quantile(x,0.999)
x_new = np.linspace(x_min, x_max, 500)

#df_xy = pd.DataFrame({"x":x, "y":y})
#df_xy.drop_duplicates(inplace=True)
```

```python
# Interpolate using splines
t,c,k = interpolate.splrep(x, y, t=q_knots, s=1)
yfit = interpolate.BSpline(t,c,k)(x_new)  
```

```python
# Plot 
# set figure size
plt.figure(figsize=(8, 6))
# set title
plt.title("Return today vs Return tomorrow")
# plot scatter
plt.plot(x, y, '.', c="grey", label="original")
# plot spline fit
plt.plot(x_new, yfit, '-', c="g", label="spline fit")
# plot legend
plt.legend(loc='best', fancybox=True, shadow=True)
# set x limits
plt.xlim(x_min,x_max)
# set x label and y label
plt.xlabel("today",size=14)
plt.ylabel("tomorrow",size=14)
# add grid
plt.grid()
# show
plt.show() 
```

# Calculate confidence intervals using bootstrap

```python
# Initialize matices to store result
Yfit = np.zeros(shape=(1000,yfit.shape[0]))
y_lower = np.zeros(shape=yfit.shape[0])
y_upper = np.zeros(shape=yfit.shape[0])
```

```python
%%time
# Create 1000 Bootstrapped samples
for i in range(1000):

    # Random sample with replacements
    df_sample = df.iloc[np.random.randint(0, len(df), size=len(df))]
    
    # Sort
    df_sample.sort_values(by="Return",ascending=True,inplace=True)
    
    # xaxis is the return today
    x = df_sample["Return"]
    # yaxis is the return tomorrow
    y = df_sample["ShiftReturn"]

    # interplate using B soline
    t,c,k = interpolate.splrep(x, y, t=q_knots, s=1)
    yfit = interpolate.BSpline(t,c,k)(x_new) 
    
    # store result
    Yfit[i,:] = yfit
```

```python
Yfit
```

```python
# Calculate upper and lower confidence interval
for i in range(yfit.shape[0]):
    y_lower[i] = np.quantile(Yfit[:,i],0.01)
    y_upper[i] = np.quantile(Yfit[:,i],0.99)
```

```python
# Plot 
plt.figure(figsize=(8, 6))
plt.title("Return today vs Return tomorrow")
plt.plot(x, y, '.', c="grey", label="original")
plt.plot(x_new, y_upper, '-', c="b", label="upper 0.99")
plt.plot(x_new, y_lower, '-', c="r", label="lower 0.01")
plt.legend(loc='best', fancybox=True, shadow=True)
plt.xlim(x_min,x_max)
plt.xlabel("today",size=14)
plt.ylabel("tomorrow",size=14)
plt.grid()
plt.show() 
```

# Anomaly detection

```python
# Resort the dataframe
df.sort_values(by="Date",ascending=True,inplace=True)
```

```python
# Calculate the moving average and standard deviation
window_size = 90
df['Moving Average'] = df["Return"].rolling(window=window_size).mean()
df['Standard Deviation'] = df['Return'].rolling(window=window_size).std()
df.dropna(inplace=True)
```

```python
# Set the threshold for anomaly detection
threshold = 4.5
df['Lower Bound'] = df['Moving Average'] - (df['Standard Deviation'] * threshold)
df['Upper Bound'] = df['Moving Average'] + (df['Standard Deviation'] * threshold)

# Identify the anomalies
df['Anomaly'] = np.where((df['Return'] > df['Upper Bound']) | (df['Return'] < df['Lower Bound']), 1, 0)
```

```python
# Calculate performance metrics for ruptures change points
from sklearn.metrics import mean_squared_error
import numpy as np

# Create binary change point indicator series
cp_indicator = np.zeros(len(test_signal))
for cp in result[:-1]:
    cp_indicator[cp] = 1

# Calculate MSE between consecutive segments
mse_list = []
prev_cp = 0
for cp in result:
    segment = test_signal[prev_cp:cp]
    if len(segment) > 1:
        mse = mean_squared_error([np.mean(segment)]*len(segment), segment)
        mse_list.append(mse)
    prev_cp = cp

print('MSE: {:3.3f}'.format(np.mean(mse_list)))

# Plot the segmentation results
plt.figure(figsize=(18, 4))
plt.plot(df.index, test_signal, alpha=0.6)

# Add vertical lines for change points and print dates
print("\nChange points detected at:")
for cp in result[:-1]:
    plt.axvline(x=df.index[cp], color='red', linestyle='--', alpha=0.8)
    print(df.index[cp].strftime('%Y-%m-%d'))

plt.ylabel("Return")
plt.title("Change Point Detection Results")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# If you need to save the segmentation information
segmentation = np.array([[cp, 0] for cp in result[:-1]])  # Format similar to original code
Date = df.index.values.astype(np.datetime64)

# Print change point dates
print("\nDetailed change points:")
for i in range(len(segmentation)):
    tmp = segmentation[i,0]
    print(f"Change point {i+1}: {Date[tmp]}")
```
